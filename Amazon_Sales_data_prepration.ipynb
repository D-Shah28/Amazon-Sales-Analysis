```python
# import libraries
import pandas as pd
import numpy as np
```

```python
# reading the sales data and storing it in data frame by using read_csv function of pandas library
data = pd.read_csv('https://raw.githubusercontent.com/D-SHAH-CU/Amazon-Sales-Analysis/main/Amazon%20Sales%20data%20(1).csv')
```

```python
# displaying the first 5 rows to understand the data
data.head()
```

```python
# check the no of rows and columns
data.shape
```

```python
# checking the datatypes of all columns
data.info()
```

```python
# here order date and ship date is in object datatype, which we have to convert into date datatype
# converting the datatype of ship date and order date
data['Order Date'] = pd.to_datetime(data['Order Date'], format='%m/%d/%Y')
data['Ship Date'] = pd.to_datetime(data['Ship Date'], format='%m/%d/%Y')
```

```python
#checking if the conversion of datatype happened or not using info()
data.info()
```

```python
data.head()
```

```python
# checking the missing values
data.isnull().sum()
```

```python
# check duplicates value
data.duplicated()
```

The data cleaning process has been successfully completed, with no missing or duplicated values identified. The primary issue addressed was the correction of the date format in the "Order Date" and "Ship Date" columns.

This cleaned dataset will now be used for further analysis and report preparation in Power BI. The data will be exported in CSV format using the to_csv function for seamless integration with Power BI.

```python
data.to_csv('amazon_sales_cleaned.csv', index=False)
```
